```python
# prompt: 上の２変量正規分布の周辺確率密度関数のグラフを作成して下さい．

# Compute the marginal probability density function for X
# Integrate the joint PDF over Y
# For the bivariate normal distribution defined by mean=[mu_x, mu_y] and covariance [[sigma_x^2, rho*sigma_x*sigma_y], [rho*sigma_x*sigma_y, sigma_y^2]],
# the marginal distribution of X is a normal distribution with mean mu_x and variance sigma_x^2.
# In our example: mean = [0.5, 0.5], cov = [[0.1, 0], [0, 0.1]]
# So, mu_x = 0.5 and sigma_x^2 = 0.1.
# The marginal PDF for X is f_X(x) = N(x | mu_x, sigma_x^2)


from scipy.stats import norm

# Parameters for the marginal distribution of X
mu_x = mean[0]
sigma_x_sq = cov[0][0]
sigma_x = np.sqrt(sigma_x_sq)

# Create a normal distribution object for X
rv_x = norm(mu_x, sigma_x)

# Evaluate the marginal PDF for X over the range of x
pdf_x = rv_x.pdf(x)

# Plot the marginal PDF for X
fig_marginal_x = plt.figure()
ax_marginal_x = fig_marginal_x.add_subplot(111)

ax_marginal_x.plot(x, pdf_x, label=f'μ={mu_x}, σ²={sigma_x_sq:.2f}')

# Set labels and title
ax_marginal_x.set_xlabel('X')
ax_marginal_x.set_ylabel('確率密度')
ax_marginal_x.set_title('Xの周辺確率密度関数 (正規分布)')

# Set limits
ax_marginal_x.set_xlim([-0.5, 1.5])
ax_marginal_x.set_ylim([0, np.max(pdf_x) * 1.1])

ax_marginal_x.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Compute the marginal probability density function for Y
# Integrate the joint PDF over X
# For the bivariate normal distribution, the marginal distribution of Y is a normal distribution with mean mu_y and variance sigma_y^2.
# In our example: mean = [0.5, 0.5], cov = [[0.1, 0], [0, 0.1]]
# So, mu_y = 0.5 and sigma_y^2 = 0.1.
# The marginal PDF for Y is f_Y(y) = N(y | mu_y, sigma_y^2)

# Parameters for the marginal distribution of Y
mu_y = mean[1]
sigma_y_sq = cov[1][1]
sigma_y = np.sqrt(sigma_y_sq)

# Create a normal distribution object for Y
rv_y = norm(mu_y, sigma_y)

# Evaluate the marginal PDF for Y over the range of y
pdf_y = rv_y.pdf(y)

# Plot the marginal PDF for Y
fig_marginal_y = plt.figure()
ax_marginal_y = fig_marginal_y.add_subplot(111)

ax_marginal_y.plot(y, pdf_y, label=f'μ={mu_y}, σ²={sigma_y_sq:.2f}', color='orange')

# Set labels and title
ax_marginal_y.set_xlabel('Y')
ax_marginal_y.set_ylabel('確率密度')
ax_marginal_y.set_title('Yの周辺確率密度関数 (正規分布)')

# Set limits
ax_marginal_y.set_xlim([-0.5, 1.5])
ax_marginal_y.set_ylim([0, np.max(pdf_y) * 1.1])

ax_marginal_y.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# prompt: A証券の投資収益率の期待値=10%，標準偏差=8%，B証券の投資収益率の期待値=14%，標準偏差=10%，A証券とB証券の投資収益率の相関係数=0.2%とする．このとき，A証券とB証券の2証券からなるポートフォリオにおいて，A証券の組み入れ比率を0%から100%まで変化させたときの，ポートフォリオ投資収益率の（標準偏差，期待値）のグラフを作成して下さい．

!pip install japanize-matplotlib

import japanize_matplotlib
import numpy as np
import matplotlib.pyplot as plt

# A証券とB証券のパラメータ
expected_return_a = 0.10    # 10%
std_dev_a = 0.08            # 8%
expected_return_b = 0.14    # 14%
std_dev_b = 0.10            # 10%
correlation_ab = 0.2      # 0.2% (注意: 相関係数は通常 -1 から 1 の範囲です。0.2%は0.002として扱います)

# 相関係数から共分散を計算
covariance_ab = correlation_ab * std_dev_a * std_dev_b

# A証券の組み入れ比率を0%から100%まで変化させる
# 100分割で計算してみる
weights_a = np.linspace(0, 1, 101)
weights_b = 1 - weights_a

# ポートフォリオの期待リターンと標準偏差を計算
portfolio_returns = weights_a * expected_return_a + weights_b * expected_return_b
portfolio_variances = (weights_a**2 * std_dev_a**2 +
                       weights_b**2 * std_dev_b**2 +
                       2 * weights_a * weights_b * covariance_ab)
portfolio_std_devs = np.sqrt(portfolio_variances)

# グラフの作成
plt.figure(figsize=(10, 6))
plt.plot(portfolio_std_devs, portfolio_returns, marker='o', markersize=4)

# グラフのタイトルとラベルを設定
plt.title('2証券ポートフォリオの効率フロンティア')
plt.xlabel('ポートフォリオの標準偏差')
plt.ylabel('ポートフォリオの期待リターン')
plt.grid(True)

# 各証券の位置に注釈を追加
plt.annotate('A証券', (std_dev_a, expected_return_a), textcoords="offset points", xytext=(10,10), ha='center')
plt.annotate('B証券', (std_dev_b, expected_return_b), textcoords="offset points", xytext=(-10,10), ha='center')
plt.plot(std_dev_a, expected_return_a, 'ro') # A証券の位置を赤丸でプロット
plt.plot(std_dev_b, expected_return_b, 'bo') # B証券の位置を青丸でプロット


# グラフを表示
plt.show()
```

```python
# prompt: ベルヌーイ分布に従う乱数を100回，1000回，10000回発生させてそれぞれの平均を求めて下さい．

import numpy as np # Import numpy library

# Set the probability of success for the Bernoulli distribution
p = 0.3 # For example, a fair coin flip

# Generate random numbers and calculate the mean for different sample sizes
sample_sizes = [100, 1000, 10000]
means = {}

for n in sample_sizes:
  # Generate Bernoulli random numbers (0 or 1)
  # np.random.binomial(n=1, p=p, size=n) generates n trials of a Bernoulli experiment
  bernoulli_samples = np.random.binomial(n=1, p=p, size=n)

  # Calculate the mean of the samples
  sample_mean = np.mean(bernoulli_samples)
  means[n] = sample_mean

  print(f'Sample size: {n}, Mean: {sample_mean:.4f}')

# You can also access the means for each sample size from the dictionary
# print(means[100])
# print(means[1000])
# print(means[10000])
```

```python
# prompt: ベルヌーイ分布に従う乱数を10000回発生させてその平均を標準化する，すなわち,標準化平均を求める．このことを1000回繰り返したときのヒストグラムを，その総面積が１になるようにして描く，さらに，標準正規分布の確率密度関数を重ねて描いて下さい．

!pip install japanize-matplotlib

import japanize_matplotlib
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the parameters for the Bernoulli distribution
p = 0.5  # Probability of success (e.g., for a fair coin flip)

# Set the number of trials for each sample
n_trials_per_sample = 10000

# Set the number of times to repeat the experiment
n_experiments = 1000

# List to store the standardized means
standardized_means = []

# Repeat the experiment n_experiments times
for _ in range(n_experiments):
    # Generate n_trials_per_sample Bernoulli random numbers
    bernoulli_samples = np.random.binomial(n=1, p=p, size=n_trials_per_sample)

    # Calculate the sample mean
    sample_mean = np.mean(bernoulli_samples)

    # Calculate the theoretical mean and variance for a Bernoulli distribution
    # Mean (E[X]) = p
    # Variance (Var[X]) = p * (1 - p)
    # For the sample mean of n independent Bernoulli trials, the mean is still p
    # and the variance is Var[X] / n = p * (1 - p) / n
    theoretical_mean_of_mean = p
    theoretical_variance_of_mean = p * (1 - p) / n_trials_per_sample
    theoretical_std_dev_of_mean = np.sqrt(theoretical_variance_of_mean)

    # Standardize the sample mean
    # Standardized mean = (sample_mean - theoretical_mean_of_mean) / theoretical_std_dev_of_mean
    if theoretical_std_dev_of_mean > 0: # Avoid division by zero if p is 0 or 1
        standardized_mean = (sample_mean - theoretical_mean_of_mean) / theoretical_std_dev_of_mean
        standardized_means.append(standardized_mean)

# Plot the histogram of the standardized means
plt.figure(figsize=(10, 6))
# density=True normalizes the histogram so the total area is 1
plt.hist(standardized_means, bins=50, density=True, alpha=0.6, color='g', label='Standardized Means Histogram')

# Overlay the probability density function of the standard normal distribution
# The standardized mean should approximately follow a standard normal distribution due to the Central Limit Theorem
x_norm = np.linspace(min(standardized_means), max(standardized_means), 100)
plt.plot(x_norm, norm.pdf(x_norm, 0, 1), 'r-', lw=2, label='Standard Normal PDF (μ=0, σ=1)')

# Set titles and labels
plt.title('標準化平均のヒストグラムと標準正規分布')
plt.xlabel('標準化平均')
plt.ylabel('確率密度')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# prompt: daily_returnsから母分散の95％信頼区間を求める

# daily_returnsはすでに計算済みとします。
# 以下のコードは、daily_returnsから母分散の95%信頼区間を求めます。

# 標本サイズ (データの個数)
# Use the numerical daily_returns_numerical DataFrame
n = len(daily_returns_numerical)

# 標本分散 (不偏分散を使用します)
# ddof=1 で不偏分散を計算
# Use the numerical daily_returns_numerical DataFrame
sample_variance = np.var(daily_returns_numerical, ddof=1)

# 信頼水準
confidence_level = 0.95

# 有意水準 (alpha)
alpha = 1 - confidence_level

# 自由度
df_chi2 = n - 1

# 信頼区間の計算に必要なカイ二乗分布の両側パーセント点
# 下側パーセント点 (alpha/2)
chi2_lower = chi2.ppf(alpha / 2, df_chi2)
# 上側パーセント点 (1 - alpha/2)
chi2_upper = chi2.ppf(1 - alpha / 2, df_chi2)

# 母分散 (σ^2) の信頼区間の下限と上限を計算
# 信頼区間は [(n-1) * s^2 / χ²_{α/2}(n-1), (n-1) * s^2 / χ²_{1-α/2}(n-1)] で与えられる
# ただし、χ²の定義によって分母が逆になる場合があるので注意が必要です。
# 一般的に、母分散の信頼区間は [(n-1)s^2 / χ²_{upper}, (n-1)s^2 / χ²_{lower}] となります。
variance_confidence_interval_lower = (df_chi2 * sample_variance) / chi2_upper*100**2
variance_confidence_interval_upper = (df_chi2 * sample_variance) / chi2_lower*100**2

print(f"標本サイズ (n): {n}")
# Access the scalar value from the Series for printing
print(f"不偏標本分散 (s^2): {sample_variance.iloc[0]:.8f}")
print(f"信頼水準: {confidence_level}")
print(f"自由度: {df_chi2}")
print(f"カイ二乗分布下側パーセント点 (α/2={alpha/2:.3f}): {chi2_lower:.4f}")
print(f"カイ二乗分布上側パーセント点 (1-α/2={1-alpha/2:.3f}): {chi2_upper:.4f}")
# Access the scalar values from the Series for printing
print(f"\n母分散 (σ^2) の{confidence_level*100}%信頼区間:")
print(f"[{variance_confidence_interval_lower.iloc[0]:.8f}, {variance_confidence_interval_upper.iloc[0]:.8f}]")

# 必要であれば、母標準偏差 (σ) の信頼区間も計算できます
# Access the scalar values from the Series before taking sqrt for clarity, though sqrt on Series would also work
std_dev_confidence_interval_lower = np.sqrt(variance_confidence_interval_lower.iloc[0])
std_dev_confidence_interval_upper = np.sqrt(variance_confidence_interval_upper.iloc[0])

print(f"\n母標準偏差 (σ) の{confidence_level*100}%信頼区間:")
print(f"[{std_dev_confidence_interval_lower:.4f}, {std_dev_confidence_interval_upper:.4f}]")
```

```python
# prompt: daily_returnsから母平均の95％信頼区間を求める

import numpy as np
from scipy.stats import t, norm

# Use the numerical daily_returns_numerical DataFrame
data = daily_returns_numerical['7203.T']

# 標本サイズ (データの個数)
n = len(data)

# 標本平均
sample_mean = np.mean(data)

# 標本標準偏差 (不偏標準偏差 s)
# ddof=1 で不偏分散を計算し、平方根をとる
sample_std = np.std(data, ddof=1)

# 信頼水準
confidence_level = 0.95

# 有意水準 (alpha)
alpha = 1 - confidence_level

# 自由度 (t分布の場合: n-1)
df_t = n - 1

# 信頼区間の計算に必要なt分布の両側パーセント点
# α/2 の上側パーセント点 (または 1 - α/2 の下側パーセント点)
# t分布の ppf (累積分布関数の逆関数) を使用
# 両側検定の場合、α/2 に対応する点を求める
# t.ppf(1 - alpha/2, df) は、確率 1 - alpha/2 となるt値を返します。
# 例えば 95% 信頼区間 (alpha=0.05) なら、alpha/2=0.025 なので、
# t.ppf(0.975, df) は上側 2.5% 点、t.ppf(0.025, df) は下側 2.5% 点を返します。
# 対称性から t.ppf(0.025, df) = -t.ppf(0.975, df) です。
# 信頼区間は [標本平均 - t値 * (s / sqrt(n)), 標本平均 + t値 * (s / sqrt(n))]
# ここでの t値 は t.ppf(1 - alpha/2, df) です。

t_critical = t.ppf(1 - alpha / 2, df_t)

# 標準誤差 (Standard Error of the Mean)
# 標準誤差 = 標本標準偏差 / sqrt(標本サイズ)
std_error = sample_std / np.sqrt(n)

# 母平均 (μ) の信頼区間の下限と上限を計算
mean_confidence_interval_lower = sample_mean - t_critical * std_error
mean_confidence_interval_upper = sample_mean + t_critical * std_error


# 結果表示
print(f"標本サイズ (n): {n}")
# Access the scalar value from the Series if sample_mean is a Series
print(f"標本平均: {sample_mean:.8f}")
# Access the scalar value from the Series if sample_std is a Series
print(f"不偏標本標準偏差 (s): {sample_std:.8f}")
# Access the scalar value from the Series if std_error is a Series
print(f"標準誤差: {std_error:.8f}")
print(f"信頼水準: {confidence_level*100}%")
print(f"自由度: {df_t}")
print(f"t分布 両側 {alpha*100}%点 (上側 α/2={alpha/2:.3f}点): {t_critical:.4f}")

print(f"\n母平均 (μ) の{confidence_level*100}%信頼区間:")
# Access the scalar values from the Series for printing
print(f"[{mean_confidence_interval_lower:.8f}, {mean_confidence_interval_upper:.8f}]")
```
