```python
# prompt: ２変量[0,1]一様分布の同時確率密度関数のグラフを作成して下さい．

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the range for x and y
x = np.linspace(-0.5, 1.5, 100)
y = np.linspace(-0.5, 1.5, 100)
X, Y = np.meshgrid(x, y)

# Define the probability density function for a uniform distribution on [0, 1] x [0, 1]
# f(x, y) = 1 for 0 <= x <= 1 and 0 <= y <= 1
# f(x, y) = 0 otherwise
Z = np.zeros_like(X)
Z[(X >= 0) & (X <= 1) & (Y >= 0) & (Y <= 1)] = 1

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Probability Density')
ax.set_title('Joint Probability Density Function of Bivariate Uniform Distribution on [0,1] x [0,1]')

# Set limits to include the relevant region
ax.set_xlim([-0.5, 1.5])
ax.set_ylim([-0.5, 1.5])
ax.set_zlim([0, 1.5])

plt.show()
```

```python
# prompt: ２変量正規分布の同時確率密度関数のグラフを作成して下さい．

from scipy.stats import multivariate_normal

# Define the mean and covariance matrix for the bivariate normal distribution
mean = [0.5, 0.5] # Example mean
cov = [[0.1, 0], [0, 0.1]] # Example covariance matrix (diagonal for independent variables)
# For a more general covariance matrix, e.g., with correlation:
# cov = [[0.1, 0.05], [0.05, 0.1]]

# Create a multivariate normal distribution object
rv = multivariate_normal(mean, cov)

# Evaluate the probability density function for each point in the meshgrid
Z_normal = rv.pdf(np.dstack([X, Y]))

# Create the 3D plot for the bivariate normal distribution
fig_normal = plt.figure()
ax_normal = fig_normal.add_subplot(111, projection='3d')

# Plot the surface
ax_normal.plot_surface(X, Y, Z_normal, cmap='viridis')

# Set labels and title
ax_normal.set_xlabel('X')
ax_normal.set_ylabel('Y')
ax_normal.set_zlabel('確率密度', rotation=90, labelpad=10)
ax_normal.set_title('2変数正規分布の同時確率密度関数')

# Set limits (adjust based on the mean and covariance for better visualization)
ax_normal.set_xlim([-0.5, 1.5])
ax_normal.set_ylim([-0.5, 1.5])
# Auto-scale z-axis based on the generated Z values, or set manually
# ax_normal.set_zlim([0, np.max(Z_normal) * 1.1])

# Adjust tick locators
ax_normal.xaxis.set_major_locator(plt.MultipleLocator(0.5))
ax_normal.yaxis.set_major_locator(plt.MultipleLocator(0.5))
# Adjust z-axis locator based on the range of Z_normal
ax_normal.zaxis.set_major_locator(plt.MaxNLocator(nbins=5))


plt.tight_layout()
plt.show()
```

```python
# prompt: 上の２変量同時確率密度関数のグラフをX軸方向から水平に見たグラフを作成して下さい．

# Create the 3D plot for the bivariate normal distribution
fig_normal_x_view = plt.figure()
ax_normal_x_view = fig_normal_x_view.add_subplot(111, projection='3d')

# Plot the surface
ax_normal_x_view.plot_surface(X, Y, Z_normal, cmap='viridis')

# Set labels and title
ax_normal_x_view.set_xlabel('X')
ax_normal_x_view.set_ylabel('Y')
ax_normal_x_view.set_zlabel('確率密度', rotation=90, labelpad=10)
ax_normal_x_view.set_title('2変数正規分布の同時確率密度関数 (X軸方向から見た図)')

# Set limits (adjust based on the mean and covariance for better visualization)
ax_normal_x_view.set_xlim([-0.5, 1.5])
ax_normal_x_view.set_ylim([-0.5, 1.5])
# Auto-scale z-axis based on the generated Z values, or set manually
# ax_normal_x_view.set_zlim([0, np.max(Z_normal) * 1.1])

# Adjust tick locators
ax_normal_x_view.xaxis.set_major_locator(plt.MultipleLocator(0.5))
ax_normal_x_view.yaxis.set_major_locator(plt.MultipleLocator(0.5))
# Adjust z-axis locator based on the range of Z_normal
ax_normal_x_view.zaxis.set_major_locator(plt.MaxNLocator(nbins=5))

# Set the view angle to be from the direction of the X-axis
# elev controls the elevation (angle above the xy plane), azim controls the azimuth (rotation in the xy plane)
# azim = 0 looks along the positive y-axis towards the origin.
# azim = 90 looks along the positive x-axis towards the origin.
# azim = -90 or 270 looks along the negative x-axis towards the origin.
# To look horizontally from the X-axis direction, we need azim to be 270 or -90.
# Adjust elev if needed, elev=0 would be directly from the side, elev=30 is a common default
ax_normal_x_view.view_init(elev=30, azim=270) # Adjust elev as desired

plt.tight_layout()
plt.show()
```

```python
# prompt: 上の２変量正規分布同時確率密度関数の２次元等高線グラフを作成して下さい．

# Create the contour plot for the bivariate normal distribution
fig_contour = plt.figure()
ax_contour = fig_contour.add_subplot(111)

# Create the contour plot using the evaluated PDF values
contour = ax_contour.contour(X, Y, Z_normal, cmap='viridis')

# Add labels to the contour lines
ax_contour.clabel(contour, inline=1, fontsize=10)

# Set labels and title
ax_contour.set_xlabel('X')
ax_contour.set_ylabel('Y')
ax_contour.set_title('2変数正規分布の同時確率密度関数 等高線')

# Set limits
ax_contour.set_xlim([-0.5, 1.5])
ax_contour.set_ylim([-0.5, 1.5])

# Add a color bar to indicate the density levels
fig_contour.colorbar(contour, label='確率密度')

plt.tight_layout()
plt.show()
```
